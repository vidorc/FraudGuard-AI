```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
credit_card_scam = pd.read_csv('/content/creditcard[1].csv')
```

```python
credit_card_scam.head()
```

```python
credit_card_scam.tail()
```

```python
credit_card_scam['Class'].value_counts()
```

```python
legit = credit_card_scam[credit_card_scam.Class == 0]
fraud = credit_card_scam[credit_card_scam.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
credit_card_scam.drop('Time', axis=1).groupby('Class').mean()
```

```python
legit_sample  = legit.sample(n=492)
```

```python
new_dataset = pd.concat([legit_sample,fraud], axis=0)
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape , X_test.shape ,X_train.shape )
```

```python
model = LogisticRegression(max_iter=1000)
```

```python
model.fit(X_train, Y_train)
```

```python
x_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(x_train_prediction, Y_train)
```

```python
print('Accuracy on Training data ; ', training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on Test data ; ', test_data_accuracy)
```
